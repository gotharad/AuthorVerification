```python
def format_pair(post1, post2):
    formatted_post = {}
    formatted_post['userId_post1'] = post1[0]
    formatted_post['index_post1'] = post1[1]
    formatted_post['userId_post2'] = post2[0]
    formatted_post['index_post2'] = post2[1]
    formatted_post['Characters_post1'] = post1[2]
    formatted_post['Characters_post2'] = post2[2]
    formatted_post['Words_post1'] = post1[3]
    formatted_post['Words_post2'] = post2[3]
    formatted_post['Captials_post1'] = post1[4]
    formatted_post['Captials_post2'] = post2[4]
    formatted_post['Sentences_post1'] = post1[5]
    formatted_post['Sentences_post2'] = post2[5]
    formatted_post['CharactersPerWord_post1'] = post1[6]
    formatted_post['CharactersPerWord_post2'] = post2[6]
    formatted_post['WordsPerSentence_post1'] = post1[7]
    formatted_post['WordsPerSentence_post2'] = post2[7]
    formatted_post['Words1_post1'] = post1[8]
    formatted_post['Words1_post2'] = post2[8]
    formatted_post['Misspellings_post1'] = post1[9]
    formatted_post['Misspellings_post2'] = post2[9]
    formatted_post['MisspellingRatio_post1'] = post1[10]
    formatted_post['MisspellingRatio_post2'] = post2[10]
    formatted_post['same_author'] = 0
    if post1[0] == post2[0]:
        formatted_post['same_author'] = 1
    return formatted_post
```

```python
def get_same_author_pairs(authors_works, N, num_posts):
  remainder = 0
  pairs = []
  for author in authors_works:
    author_works = authors_works[author]
    curr_num_posts = len(author_works)
    if curr_num_posts < 2:
      continue
    random.shuffle(author_works)
    num_pairs = N*curr_num_posts/float(num_posts)
    remainder += num_pairs % 1
    num_pairs = int(num_pairs)
    if remainder >= 1:
      num_pairs += 1
      remainder -= 1
    for i in range(num_pairs):
      pair = format_pair(author_works[2*i], author_works[2*i+1])
      pairs.append(pair)
  return pairs
```

```python
def get_different_author_pairs(authors_works, dataset, N, num_posts):
  remainder = 0
  pairs = []
  for author in authors_works:
    author_works = authors_works[author]
    curr_num_posts = len(author_works)
    random.shuffle(author_works)
    num_pairs = N*curr_num_posts/float(num_posts)
    remainder += num_pairs % 1
    num_pairs = int(num_pairs)
    if remainder >= 1:
      num_pairs += 1
      remainder -= 1
    for i in range(num_pairs):
      second_post = None
      while True:
        ind = random.randint(0,num_posts-1)
        second_post = dataset.iloc[random.randint(0,num_posts-1)]
        if second_post['userId'] != author:
          break
      second_post_formatted = (second_post['userId'], ind, second_post['Characters'], second_post['Words'], second_post['Captials'], second_post['Sentences'], second_post['CharactersPerWord'], second_post['WordsPerSentence'], second_post['Words1'], second_post['Misspellings'], second_post['MisspellingRatio'])
      pair = format_pair(author_works[i], second_post_formatted)
      pairs.append(pair)
  return pairs
```

```python
def create_author_verification_dataset(dataset, authors_works, N):
  num_posts = len(dataset)
  num_authors_one_post = (imdb.groupby(['userId'])['userId'].count() == 1).values.sum()
  pairs = get_same_author_pairs(authors_works, N, num_posts - num_authors_one_post)
  pairs = pairs + (get_different_author_pairs(authors_works, imdb, N, num_posts))
  random.shuffle(pairs)
  return pairs
```

```python
# Train multilayer perceptron on pairs_df using same_author as the label, and the other columns as features.
# PyTorch will be used to implement the model.

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

class AuthorVerificationDataset(Dataset):
    def __init__(self, df):
        self.df = df
        self.labels = df['same_author'].values
        self.features = df.drop(['same_author'], axis=1).values
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
    
class AuthorVerificationModel(nn.Module):
    def __init__(self, input_size):
        super(AuthorVerificationModel, self).__init__()
        self.fc1 = nn.Linear(input_size, 100)
        self.fc2 = nn.Linear(100, 50)
        self.fc3 = nn.Linear(50, 1)
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
    
def train_model(model, train_loader, test_loader, num_epochs, learning_rate):
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    for epoch in range(num_epochs):
        correct = 0
        total = 0
        for i, data in enumerate(train_loader):
            # if i % 100 == 0:
            #     print(i)
            inputs, labels = data
            inputs = inputs.float()
            labels = labels.float()
            labels = labels.view(-1, 1)
            optimizer.zero_grad()
            outputs = model(inputs)
            if torch.isnan(outputs).any().item():
                print('nan')
                continue
            try:
                loss = criterion(outputs, labels)
            except:
                print(outputs)
            loss.backward()
            total += labels.size(0)
            correct += (outputs == labels).sum().item()
            optimizer.step()
        if epoch % 1 == 0:
            print('Epoch: {}/{}'.format(epoch, num_epochs))
            print('Train loss: {}'.format(correct/total))
            with torch.no_grad():
                correct = 0
                total = 0
                for data in test_loader:
                    inputs, labels = data
                    inputs = inputs.float()
                    labels = labels.float()
                    outputs = model(inputs)
                    # predicted = torch.round(outputs)
                    total += labels.size(0)
                    correct += (outputs == labels).sum().item()
                print('Test accuracy: {}'.format(correct/total))
        # print('epoch')
    return model

train_df = pairs_df.sample(frac=0.8, random_state=200)
test_df = pairs_df.drop(train_df.index)

train_dataset = AuthorVerificationDataset(train_df)
test_dataset = AuthorVerificationDataset(test_df)

train_loader = DataLoader(train_dataset, batch_size=10, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=10, shuffle=True)

model = AuthorVerificationModel(22)
model = train_model(model, train_loader, test_loader, 10, 0.001)
```
