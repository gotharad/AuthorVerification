```python
imdb_train, imdb_test, ID_train, ID_test = model_selection.train_test_split(imdb1m.content, imdb1m.userId, train_size = 0.8)
ID_train = ID_train.to_list()
ID_test = ID_test.to_list()
ID_train = [1 if i == 4935810 else 0 for i in ID_train]
ID_test = [1 if i == 4935810 else 0 for i in ID_test]
```

```python
# Stopword removal, converting uppercase into lower case, and lemmatization
stopwords = nltk.corpus.stopwords.words('english')
lemmatizer = WordNetLemmatizer()
train_data_without_stopwords = []

for review in imdb_train:
    doc = re.sub('[^a-zA-Z]', ' ', review)
    doc = doc.lower()
    doc = doc.split()
    doc = [lemmatizer.lemmatize(word) for word in doc if not word in set(stopwords)]
    doc = ' '.join(doc)
    train_data_without_stopwords.append(doc)

```

```python
train_vectorizer = TfidfVectorizer() 
train_vectors = train_vectorizer.fit_transform(train_data_without_stopwords)
clf = LogisticRegression()
clf.fit(train_vectors, ID_train)
y_train_pred = clf.predict(train_vectors) 

print(metrics.accuracy_score(ID_train, y_train_pred))
print(metrics.f1_score(ID_train, y_train_pred, average = "weighted"))
```

```python
test_data_without_stopwords = []
for review in imdb_test:
    doc = re.sub('[^a-zA-Z]', ' ', review)
    doc = doc.lower()
    doc = doc.split()
    doc = [lemmatizer.lemmatize(word) for word in doc if not word in set(stopwords)]
    doc = ' '.join(doc)
    test_data_without_stopwords.append(doc)

X_test_vect = train_vectorizer.transform(test_data_without_stopwords)
y_pred = clf.predict(X_test_vect)




```

```python
print(metrics.accuracy_score(ID_test, y_pred))
print(metrics.f1_score(ID_test, y_pred, average = "weighted"))
```
